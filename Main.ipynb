```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from graph_generator import generate_graph
from graph_sparsifier import graph_sparsification
```

### Initialize Graph
Randomly generate the demand graph using the gravity model

```python
np.random.seed(3)
n_nodes = 9
node_population = 1000*(1-np.random.power(50, n_nodes))
np.random.shuffle(node_population)
fmin = 1

node_coord = [[np.random.uniform(0,40), np.random.uniform(0,40)] for i in range(0,n_nodes)]

# Generate graph
G_OD_full = generate_graph(node_coord, node_population, fmin)

# Get the edge properties with correct order
edges = G_OD_full.edges()
od_array = np.array([G_OD_full[u][v]['OD_Pax'] for u, v in edges])

plt.subplots(figsize = (10, 7))
nx.draw(G_OD_full, node_coord, node_size = 10*node_population,  width = od_array/10)
```

### Solve the OD to LEG problem:

```python
leg_array = graph_sparsification(G_OD_full)
plt.subplots(figsize = (10, 7))
nx.draw(G_OD_full, node_coord, node_size = 10*node_population,  width = leg_array/10)
```
